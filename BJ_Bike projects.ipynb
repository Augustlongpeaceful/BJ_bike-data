```python
## Here we directly import the file we have produced before
df_od = pd.read_csv(r'C:\Users\18369\Desktop\thesis ref\Work_RA\F_assistant\O_D_trips.csv', index_col=0)

df_list = pd.read_csv(r'C:\Users\18369\Desktop\thesis ref\Work_RA\F_assistant\List.csv', index_col=0)
```

```python
## Select the final reference column for both csv table
od = df_od['O_D_trips'] 
ls = df_list['O_D_ref']
```

```python
## DataFrame isin function helps to check whether or not the O_D_trips is really (NOT) a subset of O_D_ref
result1 = ~od.isin(ls) ## ~ is not
result1.head(17)
```

```python
## transfer bool intt 1 & 0 
result2 = result1.astype(int)
result2
```

```python
## convert 1 & 0 into object in a dataframe
result3 = pd.DataFrame(result2)
result3.head(30)
```

```python
## Add this result to the final column of the original table, then is done
df_od['Result'] = result3
df_od.head(17)
```
